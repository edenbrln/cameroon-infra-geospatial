```python
import geopandas as gpd
import pandas as pd

# 분석 결과 + 구역 경계 합치기
results = pd.read_csv('../data/processed/final_analysis_results.csv')
districts = gpd.read_file('../data/processed/douala_yaounde_districts.geojson')

# geometry 붙이기
gdf_final = districts.merge(results, on=['district_name', 'city'])

print(gdf_final.columns.tolist())
print(f"구역 수: {len(gdf_final)}")

# Tableau가 읽을 수 있는 GeoJSON으로 저장
gdf_final.to_file('../data/processed/tableau_infra_analysis.geojson', driver='GeoJSON')
print("Tableau용 GeoJSON 저장 완료!")
```
